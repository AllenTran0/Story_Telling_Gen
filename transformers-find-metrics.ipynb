```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
model = "/Users/allentran/Downloads/archive"

tokenizer = AutoTokenizer.from_pretrained(model)
model = AutoModelForCausalLM.from_pretrained(model).to(device)

model
```

```python
from datasets import load_dataset

dataset = load_dataset("roneneldan/TinyStories")

dataset
```

```python
import evaluate

bleu = evaluate.load("bleu")
```

```python
for i in range(10):
    references = [dataset['train'][slice(i, i + 1, None)]]

    print(references[0]['text'])
```

```python
for i in range(10):
    references = [dataset['train'][slice(i, i + 1, None)]]

    store = references[0].get('text')[0].split(".")

    prompt = store[0]

    

    print(references[0]['text'][0])

    inputs = tokenizer(prompt, return_tensors = "pt").to(device)
    outputs = model.generate(inputs.input_ids, max_new_tokens = len(references[0]['text'][0]), do_sample = True, top_k = 50).to(device)


    output_str = tokenizer.batch_decode(outputs)
    print(output_str[0])

    results = bleu.compute(predictions = output_str,references = references[0]['text'])

    print(results)
```
